```python
from google.colab import files
import pdfplumber
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

TECH_SKILLS = {
    "python", "c++", "java", "sql", "html", "css", "javascript", "react", "node.js",
    "machine learning", "deep learning", "artificial intelligence", "data science",
    "big data", "mongodb", "mysql", "docker", "git", "kubernetes", "tensorflow",
    "pytorch", "pandas", "numpy", "scikit-learn", "matplotlib", "seaborn", "flask",
    "django", "nlp", "excel", "power bi", "tableau", "azure", "aws", "gcp"
}

def extract_text_from_pdf(file_path):
    with pdfplumber.open(file_path) as pdf:
        return "\n".join([page.extract_text() for page in pdf.pages if page.extract_text()])

def extract_skills(text):
    doc = nlp(text.lower())
    found_skills = set()

    for token in doc:
        if token.text in TECH_SKILLS:
            found_skills.add(token.text)

    for skill in TECH_SKILLS:
        if " " in skill and skill in text.lower():
            found_skills.add(skill)

    return sorted(found_skills)

# File Upload
uploaded = files.upload()

for filename in uploaded.keys():
    print(f"\n📄 Analyzing: {filename}")
    text = extract_text_from_pdf(filename)
    skills = extract_skills(text)
    print("\n🧠 Extracted Skills:\n", ", ".join(skills) if skills else "No known technical skills found.")
```
